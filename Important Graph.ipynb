```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,10)
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)
y5=np.log(x)
```

```python
plt.figure(figsize=(16,12))
plt.subplot(2,5,1)
plt.plot(x,y1,'ro ')
plt.title('y1=x')
plt.subplot(2,5,2)
plt.plot(x,y2,'b.-')
plt.title('y2=x**2')
plt.subplot(2,5,3)
plt.plot(x,y3,'y*')
plt.title('y3=x**3')
plt.subplot(2,5,4)
plt.plot(x,y4,'g^')
plt.title('y4=sqrt(x)')
plt.subplot(2,5,5)
plt.plot(x,y5,'ks')
plt.title('y5=logx')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(5,1,1)
plt.plot(x,y1,'ro ')
plt.title('y1=x')
plt.subplot(5,1,2)
plt.plot(x,y2,'b.-')
plt.title('y2=x**2')
plt.subplot(5,1,3)
plt.plot(x,y3,'y*')
plt.title('y3=x**3')
plt.subplot(5,1,4)
plt.plot(x,y4,'g^')
plt.title('y4=sqrt(x)')
plt.subplot(5,1,5)
plt.plot(x,y5,'ks')
plt.title('y5=logx')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees
plt.show()
```
